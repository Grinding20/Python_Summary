```python
# 1
def jiecheng_2(n):
    if n == 1:
        return 1
    else:
        return  n+jiecheng_2(n-1)

def qiuhe(n):
    if n == 1:
        return jiecheng_2(1)
    else:
        return jiecheng_2(n)+ qiuhe(n-1)
print(qiuhe(int(input('请输入一个数字'))))
```

```python
# 2
fuhao = input('请输入一个符号：')
hangshu = int(input('请输入一个行数：'))
def tixing(h,f):
    for i in range(h):
        print('  '*(h-i-1),(f+' ')*(i+1),f*h,(' '+f)*(i+1))
    
    return  ' '
print(tixing(hangshu,fuhao))
```

```python
# 5(a)统计A的词与词频
import random
A=[]
for i in range(100000):
    a=random.randint(1,10000)
    A.append(a)
numbers = A
numbers_dict_1={}
for num in numbers:
    if num in numbers_dict_1:
        numbers_dict_1 [num] += 1
    else:
        numbers_dict_1 [num] = 1
A_list = []        
for key, value in numbers_dict_1.items():  
    a = key, value
    A_list.append(a) 
```

```python
# 5(a)统计B的词与词频
import random
B=[]
for i in range(100000):
    b=random.randint(5000,15000)
    B.append(b)
numbers = B
numbers_dict_2={}
for num in numbers:
    if num in numbers_dict_2:
        numbers_dict_2 [num] += 1
    else:
        numbers_dict_2 [num] = 1
B_list = []        
for key, value in numbers_dict_2.items():  
    c = key, value
    B_list.append(c) 
```
